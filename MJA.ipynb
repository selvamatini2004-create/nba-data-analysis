```python
def get_team_data(url ) :

    team_page_source = rq.get( 'https://www.basketball-reference.com' + url ,headers = header)
    team_page_source.encoding = 'utf-8'
    team_soup = bs(team_page_source.text , 'html.parser')
    
    #gets the candidates names 
    player_list = []
    names = team_soup.select('table[id^="mvp"] a[href^="/players/"]')
    for name in names :
        player_list.append(name.get_text())

    #gets the year of the season 
    season = team_soup.find('h1').get_text().split(' ')[0]

    return season , player_list
```
