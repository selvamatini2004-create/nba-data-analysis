```python
import requests as rq
from bs4 import BeautifulSoup as bs 
from bs4 import Comment
from random_user_agent.user_agent import UserAgent
from random_user_agent.params import SoftwareName, OperatingSystem
import pandas as pd
import time
```

```python
software_names = [SoftwareName.CHROME.value]
operating_systems = [OperatingSystem.WINDOWS.value, OperatingSystem.LINUX.value]   
user_agent_rotator = UserAgent(software_names=software_names, operating_systems=operating_systems, limit=100)
user_agents = user_agent_rotator.get_user_agents()
user_agent = user_agent_rotator.get_random_user_agent()
header = {"User-Agent" : user_agent , "Accept-Language": "en-US"}
```

```python
page_source = rq.get('https://www.basketball-reference.com/playoffs/', headers = header)
soup = bs(page_source.text, 'html.parser')
teams_href = [a['href']for a in soup.select('td[data-stat^="champion"] a[href^="/teams/"]')]
```

```python
def get_team_data(url) : 
    
    #gets the data for the champion
    team_page_source = rq.get( 'https://www.basketball-reference.com' + url ,headers = header)
    team_soup = bs(team_page_source.text , 'html.parser')

    #gets the seasons years 
    season = team_soup.select_one('h1').text.split('\n')[1]

    #the champion name using  
    champion_name = team_soup.select_one('h1').text.split('\n')[2]
    fr_link = team_soup.select('li.index a[href^="/teams/"]')
    fr_link = fr_link[0]['href']
    team_page_source2 = rq.get( 'https://www.basketball-reference.com' + fr_link ,headers = header)
    team_soup2 = bs(team_page_source2.text , 'html.parser')
    franchise_name = team_soup2.select_one('h1').text
    champion_name = franchise_name[1:-2]
    
    #champion team players names list 
    player_list = []
    players = team_soup.select('td[data-stat^="player"] a')
    for player in players:
        player_list.append(player.get_text())
        
    return season , champion_name , player_list 
```

```python
df = pd.DataFrame(columns = ['season' , 'champion_name', 'player_list'])

row_number = 1

for champion in teams_href :
    season , champion_name , player_list = get_team_data(champion)
    row = {'season' : season , 'champion_name' : champion_name , 'player_list' : player_list} 
    time.sleep(3)
    df.loc[len(df)] = row 
    print(f'row {row_number} done!')
    row_number = row_number + 1
```
